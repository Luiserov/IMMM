```python
import matplotlib.pyplot as plt

poblacion = INM.loc[:, ['Población total', 'Nombre de la entidad', 'Nombre del municipio', 'Grado de marginación, 2020']]
poblacion['porcentaje_poblacion'] = poblacion['Población total'] / poblacion.groupby('Nombre de la entidad')['Población total'].transform('sum')
porcentaje_marginacion = poblacion.groupby(['Nombre de la entidad', 'Grado de marginación, 2020'])['porcentaje_poblacion'].sum().unstack()
fig, ax = plt.subplots(figsize=(12, 4))
porcentaje_marginacion.plot(kind='bar', stacked=True, ax=ax)
ax.set_xlabel('Estado')
ax.set_ylabel('Porcentaje de población')
ax.set_title('Porcentaje de municipios por estado y grado de marginación')
ax.legend(title='Grado de marginación', bbox_to_anchor=(1, 1), loc='upper left')

plt.savefig("Porcentaje de municipios por estado y grado de marginación.png")
```

```python
marg_por_estado = INM.loc[:, ['Nombre de la entidad', 'Nombre del municipio', 'Grado de marginación, 2020']]
marg_por_estado

porcentajes = marg_por_estado.groupby('Nombre de la entidad')['Grado de marginación, 2020'].value_counts(normalize=True).mul(100).unstack()
porcentajes

ax = porcentajes.plot(kind='bar', stacked=True, figsize=(12, 6))
ax.legend(title='Grado de Marginación', bbox_to_anchor=(1, 1))
ax.set_xlabel('Estado')
ax.set_ylabel('Porcentaje de Grado de Marginación')
ax.set_title('Porcentaje de Grado de Marginación por Estado')

plt.savefig("Porcentaje grado marginación por estado.jpg")
```

## Observaciones: 

Algunas entidades cómo Chiapas, Guerrero, Oaxaca, San Luis Potosí y Yucatán tienen un porcentaje de municipios el alto grado de marginación aparentemente bajo. Pero al convertirlos a porcentaje del total de población, nos damos cuenta que es un porcentaje importante. Ésto me dice que tales municipios con alto grado de marginación, son municipios grandes. 
Igualmente, Chiapas y Campeche, tienen un porcentaje de municipios con Muy bajo grado de marginación que da una idea errónea. Al ver el porcentaje de la población total que representa, nos damos cuenta que es mucho menos. Osea que esos municipios son pequeños en relación al total de la población (menos poblados). 

##Cómo podríamos analizar que variable tiene mayor correlación con el porcentaje de analfabetismo? 

Podemos usar el método .corr() sobre nuestra base de datos. 

En éste caro, la variable de mayo correlación es el índice de marginación. 

```python
import seaborn as sns

correlation_matrix = INM.corr()

# Visualizar la matriz de correlación como un mapa de calor
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()

# Encontrar la variable con la mayor correlación con otra
max_correlation = correlation_matrix.abs().idxmax()
print('Variable con mayor correlación:')
print(max_correlation)
```
